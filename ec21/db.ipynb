```python
import oracledb
import pandas as pd

# Oracle 데이터베이스 연결 정보
username = 'admin'
password = 'DIMA3project'
dsn = 'your_dsn'  # 예: 'loca`~lhost:1521/orclpdb1'
```

```python

# Oracle 데이터베이스 연결
connection = oracledb.connect(user=username, password=password, dsn=dsn)
cursor = connection.cursor()

# CSV 파일 읽기
file_path = 'your_file_path.csv'
df = pd.read_csv(file_path)

# Oracle에 데이터 삽입하기 위한 SQL 준비
# 테이블 이름과 컬럼 이름을 확인하고 맞게 수정하세요.
insert_query = """
    INSERT INTO your_table_name (column1, column2, column3)
    VALUES (:1, :2, :3)
"""

# DataFrame의 각 행을 Oracle에 삽입
for row in df.itertuples(index=False, name=None):  # name=None removes the need for accessing attributes, allowing direct tuple usage
    cursor.execute(insert_query, row)

# 변경 사항 커밋
connection.commit()

# 연결 종료
cursor.close()
connection.close()
```
